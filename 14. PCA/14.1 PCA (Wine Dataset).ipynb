#### Split the Dataset in to Traning Set and Testing Set 

```python
from sklearn.model_selection import train_test_split 
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.25,random_state=1) 
```

#### Standerdizing the Dataset 

```python
from sklearn.preprocessing import StandardScaler 
```

```python
sc = StandardScaler() 
```

```python
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test) 
```

#### Applaying PCA 

```python
from sklearn.decomposition import PCA 
```

```python
pca = PCA(n_components = 3)
x_train = pca.fit_transform(x_train)
x_test = pca.transform(x_test) 
explained_variance = pca.explained_variance_ratio_ 
```

```python
print('explained_variance = ',explained_variance) 
```

#### Plot the explained variances

```python
features = range(pca.n_components_)
plt.bar(features, pca.explained_variance_ratio_, color='black')
plt.xlabel('PCA features')
plt.ylabel('variance %')
plt.xticks(features)
```

#### PCA for Visualization 

#### Eigendecomposition

```python
eigen_vals, eigen_vecs = np.linalg.eigh(P)

print('\nEigenvalues: \n%s' % eigen_vals)
```

```python
tot = sum(np.abs(eigen_vals))
var_exp = [(i / tot) for i in sorted(np.abs(eigen_vals), reverse=True)]
cum_var_exp = np.cumsum(var_exp)

plt.bar(range(1, eigen_vals.size + 1), var_exp, alpha=0.5, align='center',
        label='Individual')
plt.step(range(1, eigen_vals.size + 1), cum_var_exp, where='mid',
         label='Cumulative')
plt.ylabel('Explained variance ratio')
plt.xlabel('Principal components')
plt.legend(loc='best')
plt.tight_layout()
plt.show()
```

### K Means Clustering 

```python
from sklearn.cluster import KMeans 
```

```python
wcss = [] 

for i in range (1,11):
    kmeans = KMeans(n_clusters=i, init = 'k-means++', random_state=0)
    kmeans.fit(x_train), 
    wcss.append(kmeans.inertia_) 
plt.plot(range(1,11),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('wcss')
plt.show() 
```

#### Fitting KMeans Algorithm on the Dataset 

```python
kmeans1 = KMeans(n_clusters=3, init='k-means++', random_state=0) 
```

```python
y1_kmeans = kmeans1.fit_predict(x_test)  
```

```python
y1_kmeans 
```

 #### Visualization of Clusters Created in KMeans Algorithm 

```python
plt.scatter(x_test[y1_kmeans == 0,0],x_test[y1_kmeans==0,1],s=100,c='red',label='cluster1') 

plt.scatter(x_test[y1_kmeans == 1,0],x_test[y1_kmeans==1,1],s=100,c='blue',label='cluster2') 

plt.scatter(x_test[y1_kmeans == 2,0],x_test[y1_kmeans==2,1],s=100,c='green',label='cluster3')

plt.scatter(kmeans1.cluster_centers_[:,0],kmeans1.cluster_centers_[:,1],s=300,c='yellow',
            label='centroid') 
plt.title('Clusters of Customers')
plt.legend()
plt.show()
```
